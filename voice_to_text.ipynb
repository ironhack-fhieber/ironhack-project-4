```python
import speech_recognition as sr
from pydub import AudioSegment
```

## Helper methods

```python
def transcribe_audio(audio) -> str:
    r = sr.Recognizer()
    return r.recognize_google(audio, language='en')
```

```python
def convert_audio_to_text(audio_file) -> str:
    with sr.AudioFile(audio_file) as source:
        audio_data = sr.Recognizer().record(source)
        text = transcribe_audio(audio_data)
    return text
```

## Converter

```python
# define wav sound file and load it in AudioSegment
wav_file = 'test_files/record.wav'
sound = AudioSegment.from_file(wav_file)

# define path and convert wav to a pcm_s16le wav file
pcm_path = "test_files/record_pcm.wav"
sound.export(pcm_path, format="wav", codec="pcm_s16le")

# get text from voice
with sr.AudioFile(pcm_path) as source:
    audio_data = sr.Recognizer().record(source)
    text = transcribe_audio(audio_data)
text
```
